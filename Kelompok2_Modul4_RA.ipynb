```python
import pandas as pd

# Contoh data
data = pd.read_csv('/content/Copy of HARGA RUMAH JAKSEL.csv', sep=';')

# Menampilkan DataFrame
print("DataFrame awal:")
print(data)

# Feature Engineering: Ekstraksi fitur baru - Umur Rumah
if 'Tahun Pembangunan' in data.columns:
    data['Umur Rumah'] = 2024 - data['Tahun Pembangunan']
else:
    print("Kolom 'Tahun Pembangunan' tidak ditemukan dalam DataFrame.")

# Menampilkan DataFrame setelah feature engineering
print("\nDataFrame setelah feature engineering (Umur Rumah):")
print(data)
```
