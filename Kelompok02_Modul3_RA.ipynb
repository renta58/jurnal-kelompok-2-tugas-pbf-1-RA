```python
import pandas as pd

data = {
    "Bulan": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober","November","Desember"],
    2023: [357.9, 211.7, 240.6, 147.9, 219.4, 76.1, 102.5, 7.8, 2.0, 4.00, 73.80, 118.90]
}

df = pd.DataFrame(data)
miu = df[2023].mean()
sigma = df[2023].std()

display(df)

print("Dari data curah hujan pada tahun 2023, dapat diketahui besar ;")
print("Miu (Mean):", miu,"mm/bulan")
print("Sigma (Standar Deviasi):", sigma,"mm/bulan")
```

```python
import math
import random

def normal_distribution_generator(miu=0, sigma=1):
    while True:
        u1, u2 = random.random(), random.random()
        z0 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
        yield miu + z0 * sigma

# masukkan nilai miu dan sigma yang telah didapatkan
miu = (miu)
sigma = (sigma)

# membuat generator dengan nilai miu dan sigma yang telah ditentukan
sample_generator = normal_distribution_generator(miu, sigma)

print("Selamat Datang Di Aplikasi Sederhana, Memprediksi besarnya curah hujan dalam beberapa bulan mendatang:)")
answer = input("\nBerapa bulan prediksi besar curah hujan yang anda butuhkan? (1-12 bulan) ")
print("\nBaiklah ini prediksi curah hujan yang anda butuhkan ;")
#memprediksi curah hujan selama 3 bulan kedepan

for i in range(1, int(answer) + 1):
    print("Prediksi bulan ke ", i,"sebesar", next(sample_generator), "mm/bulan")
```
