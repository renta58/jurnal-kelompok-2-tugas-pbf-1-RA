```python
# Membuat fungsi filter otomatis
def filter_data(df, **kwargs):
    filtered_df = df.copy()
    for column, value in kwargs.items():
        filtered_df = filtered_df[filtered_df[column] == value]
    return filtered_df

# Contoh penggunaan sistem filter
filtered_data = filter_data(df, Gender='M', Location='California', Product_Category='Nest-USA')
print(filtered_data)
```
